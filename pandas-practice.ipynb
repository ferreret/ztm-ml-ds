```python
car_sales_missing_data = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/car-sales-extended-missing-data.csv")
```

```python
car_sales_missing_data
```

```python
car_sales_missing_data["Odometer (KM)"].fillna(car_sales_missing_data["Odometer (KM)"].mean(), inplace=True)
```

```python
car_sales_missing_data
```

```python
car_sales_missing_data.dropna(inplace=True)
```

```python
car_sales_missing_data
```

```python
car_sales_missing_data["Seats"] = 5
```

```python
car_sales_missing_data
```

```python
import random as rnd
# list random numbers between 1.3 and 4.5
random_numbers = [rnd.uniform(1.3, 4.5) for i in range(815)]

car_sales_missing_data["Engine Size"] = random_numbers
    

```

```python
car_sales_missing_data
```

```python
car_sales_missing_data["Price per Km"] = car_sales_missing_data["Price"]/car_sales_missing_data["Odometer (KM)"]
```

```python
car_sales_missing_data
```

```python
car_sales_missing_data.drop(["Price per Km"], axis=1, inplace=True)
car_sales_missing_data
```

```python
car_sales_missing_data_shuffled = car_sales_missing_data.sample(frac=1)
car_sales_missing_data_shuffled
```

```python
# Reset the indexes of the shuffled DataFrame
car_sales_missing_data_shuffled.reset_index(drop=True, inplace=True)
car_sales_missing_data_shuffled
```

```python
car_sales_missing_data_shuffled["Odometer (KM)"] = car_sales_missing_data_shuffled["Odometer (KM)"].apply(lambda x: x /1.61)
car_sales_missing_data_shuffled
```

```python
# Change column name
car_sales_missing_data_shuffled.rename(columns={"Odometer (KM)": "Odometer (Miles)"}, inplace=True)
car_sales_missing_data_shuffled
```
