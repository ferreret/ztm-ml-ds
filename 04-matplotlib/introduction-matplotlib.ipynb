## Matplotlib example workflow

```python
# 0. import matplot lib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot", xlabel="X-axis", ylabel="Y-axis")

# 5. Save & show (you save the whole figure)
# fig.savefig("simple_plot.png")
```

## Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from a dictionary
nut_butter_prices = {"Almond butter": 10, "Peanut butter": 8, "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut butter prices", xlabel="Nut butter", ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), nut_butter_prices.values());
```

```python
# Make some data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot for each different subplot
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```
