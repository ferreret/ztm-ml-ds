## Matplotlib example workflow

```python
# 0. import matplot lib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot", xlabel="X-axis", ylabel="Y-axis")

# 5. Save & show (you save the whole figure)
# fig.savefig("simple_plot.png")
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot for each different subplot
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot for each different subplot
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one shoud you use (pyplot vs matplotlib OO method?)

* When plotting something quickly, okay to use pyplot method.
* When plotting something more complex, use OO method.

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
# Pyplot method
over_50.plot(kind="scatter", x="age", y="chol", c="target", figsize=(10,6));
```

```python
# OO method midex with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
# ax.set_xlim([45, 100])
```

```python
# OO method from scratch
fig, (ax0, ax1)  = plt.subplots(nrows=2, ncols=1, figsize=(10, 12))

# fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot the data
scatter0 = ax0.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"])
# Customize the plot
ax0.set(title="Heart Disease And Cholesterol Levels",
       xlabel="Age",        
       ylabel="Cholesterol Level (mg/dL)");

# Add a legend
ax0.legend(*scatter0.legend_elements(), title="target");

# Add a horizontal line
ax0.axhline(y=over_50["chol"].mean(), 
           color="lightblue", 
           linestyle="--");

scatter1 = ax1.scatter(over_50["age"],
                       over_50["thalach"],
                       c=over_50["target"]);

ax1.set(title="Heart Disease And Maximum Heart Rate",
        xlabel="Age", 
        ylabel="Maximum Heart Rate (bpm)");

ax1.legend(*scatter1.legend_elements(), title="target");

# Add a horizontal line
ax1.axhline(y=over_50["thalach"].mean(),
            color="lightblue",
            linestyle="--");

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
plt.style.use("seaborn-whitegrid")

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"],
                     cmap="winter")
# Customize the plot
ax.set(title="Heart Disease And Cholesterol Levels",
       xlabel="Age",        
       ylabel="Cholesterol Level (mg/dL)");

# Add a legend
ax.legend(*scatter.legend_elements(), title="target");

# Add a horizontal line
ax.axhline(y=over_50["chol"].mean(),            
           linestyle="--");
```

```python
# Customizing the y and x axis
# OO method from scratch
fig, (ax0, ax1)  = plt.subplots(nrows=2, ncols=1, figsize=(10, 12))

# Plot the data
scatter0 = ax0.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"],
                     cmap="winter")
# Customize the plot
ax0.set(title="Heart Disease And Cholesterol Levels",
       xlabel="Age",        
       ylabel="Cholesterol Level (mg/dL)");
ax0.set_xlim([50, 80])

# Add a legend
ax0.legend(*scatter0.legend_elements(), title="target");

# Add a horizontal line
ax0.axhline(y=over_50["chol"].mean(),            
           linestyle="--");

scatter1 = ax1.scatter(over_50["age"],
                       over_50["thalach"],
                       c=over_50["target"],
                       cmap="winter");

ax1.set(title="Heart Disease And Maximum Heart Rate",
        xlabel="Age", 
        ylabel="Maximum Heart Rate (bpm)");
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
ax1.legend(*scatter1.legend_elements(), title="target");

# Add a horizontal line
ax1.axhline(y=over_50["thalach"].mean(),            
            linestyle="--");

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

```
